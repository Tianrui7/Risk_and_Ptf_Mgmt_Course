```python
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plt
import math
#import statsmodels
import statsmodels.api as sm
```

```python
data = pd.read_excel("NYU_Spring_2019_assignment_4.xlsx")
data.head()
```

# Question 2, replicate table 2

```python
data1 = data.copy()
# make a copy and only modify data1, not the original data
data1 = data1[data1["Month"]>pd.to_datetime("7/01/1963")]
data1 = data1[data1["Month"]<pd.to_datetime("1/01/2014")]
data1.head()
```

```python
data1 = sm.add_constant(data1)
# should build six OLS models
# RM-RF ~ SMB + HML + RMW + CMA + UMD
# SMB ~ RM-RF + HML + RMW + CMA + UMD
# HML ~ RM-RF + SMB + RMW + CMA + UMD
# RMW ~ RM-RF + SMB + HML + CMA + UMD
# CMA ~ RM-RF + SMB + HML + RMW + UMD
# UMD ~ RM-RF + SMB + HML + RMW + CMA


X1 = ['const','SMB', 'HML', 'RMW', 'CMA', 'UMD']
X2 = ['const','Mkt-RF', 'HML', 'RMW','CMA', 'UMD']
X3 = ['const','Mkt-RF', 'SMB', 'RMW', 'CMA', 'UMD']
X4 = ['const','Mkt-RF', 'SMB', 'HML','CMA', 'UMD']
X5 = ['const','Mkt-RF', 'SMB', 'HML','RMW', 'UMD']
X6 = ['const','Mkt-RF', 'SMB', 'HML' ,'RMW', 'CMA']
reg1 = sm.OLS(endog=data1['Mkt-RF'], exog=data1[X1]).fit()
reg2 = sm.OLS(endog=data1['SMB'], exog=data1[X2]).fit()
reg3 = sm.OLS(endog=data1['HML'], exog=data1[X3]).fit()
reg4 = sm.OLS(endog=data1['RMW'], exog=data1[X4]).fit()
reg5 = sm.OLS(endog=data1['CMA'], exog=data1[X5]).fit()
reg6 = sm.OLS(endog=data1['UMD'], exog=data1[X6]).fit()

# intialize an 2D matrix with nan
display_mat = np.empty((6,8))
display_mat[:] = np.nan
display_df = pd.DataFrame(display_mat)
display_df.columns = ['const', 'Mkt-RF','SMB', 'HML', 'RMW', 'CMA', 'UMD', r'$R^2$']
display_df.index = ['Mkt-RF','SMB', 'HML', 'RMW', 'CMA', 'UMD']

display_df.loc[['Mkt-RF'],['const','SMB', 'HML', 'RMW', 'CMA', 'UMD',r'$R^2$']] = np.append(np.array(reg1.params), reg1.rsquared)
display_df.loc[['SMB'],['const','Mkt-RF', 'HML', 'RMW','CMA', 'UMD',r'$R^2$']]  = np.append(np.array(reg2.params), reg2.rsquared)
display_df.loc[['HML'],['const','Mkt-RF', 'SMB', 'RMW', 'CMA', 'UMD',r'$R^2$']] = np.append(np.array(reg3.params), reg3.rsquared)
display_df.loc[['RMW'],['const','Mkt-RF', 'SMB', 'HML','CMA', 'UMD',r'$R^2$']]  = np.append(np.array(reg4.params), reg4.rsquared)
display_df.loc[['CMA'],['const','Mkt-RF', 'SMB', 'HML','RMW', 'UMD',r'$R^2$']]  = np.append(np.array(reg5.params), reg5.rsquared)
display_df.loc[['UMD'],['const','Mkt-RF', 'SMB', 'HML' ,'RMW', 'CMA',r'$R^2$']] = np.append(np.array(reg6.params), reg6.rsquared)


# since in the table 2 from the paper, intercepts are given in an annual basis
# simply multiply the data by 12 
display_df['const'] *=12
display_df
```

They don't match the previous tables exactly. Observe that the most obvious differece is that the market delivers lower excess return, which drops from about 10% to 7%.

One possible reason is that under different time periods, the performance of stocks are dominated by different factors, i.e. the "true theoratical" model has changed. Another possible reason is that every regression we do is a biased (w.r.t. time) sample of the theoratical model. It might be that different samples give us different bias.
