# Question 1, transform the data

### 1, Take the natural logarithm of the S&P500 price

```python
newdf = pd.DataFrame(raw_data["Month"])
newdf["log_S&P"]=np.log(raw_data["S&P500"])
plt.subplot(1,2,1)

plt.plot(np.linspace(1880, 2018, len(raw_data)), raw_data["S&P500"])
plt.title("S&P 500")


plt.subplot(1,2,2)
plt.plot(np.linspace(1880, 2018, len(raw_data)), newdf["log_S&P"])
plt.title("Log S&P 500")

plt.show()
```

S&P 500 is a trend variable. It naturally goes far up as time goes on. Taking log would reduce the size effect, as large values toward the end are drastically different from the tiny values at the beginning. Also, we want to measure the rate of economic development, rather than the absolute value of the development. This goal is better handled using log. 

### 2, Compute the dividend yield as dividend divided by price

```python
newdf["yield"] = raw_data["Dividends"] / raw_data["S&P500"]
plt.subplot(1,2,1)

plt.plot(np.linspace(1880, 2018, len(raw_data)), raw_data["Dividends"])
plt.title("Dividends")


plt.subplot(1,2,2)
plt.plot(np.linspace(1880, 2018, len(raw_data)), newdf["yield"])
plt.title("yield")

plt.show()
```

As the economy develops and as inflation goes upward, dividends will inevitably grow as time goes on. Hence observing the rate of return, i.e. yield, would be better than simply observing dividends.

# Question 2
Graph each of the transformed series and look at the partial autocorrelations out to lag 32. Comment briefly on features you see in the graphs, and how these relate to the partial autocorrelations.

```python
plt.figure(figsize=(18,15))
plt.subplot(3,3,1)
plt.plot(newdf["Month"], newdf["log_S&P"])
plt.title("$\log(S&T)$")

plt.subplot(3,3,2)
plt.plot(newdf["Month"], newdf["yield"])
plt.title("Yield")

plt.subplot(3,3,3)
plt.plot(newdf["Month"], newdf["log_earnings"])
plt.title("log(earnings)")

plt.subplot(3,3,4)
plt.plot(newdf["Month"], newdf["log_CPI"])
plt.title("log(CPI)")

plt.subplot(3,3,5)
plt.plot(newdf["Month"], newdf["10yr_rate"])
#plt.yticks([0,100, 200, 300, 400, 500, 600],["0%","1%", "2%", "3%", "4%", "5%", "6%"])
plt.title("10 Year Rate")

plt.subplot(3,3,6)
plt.plot(newdf["Month"], newdf["log_gold"])
plt.title("log(gold)")

plt.subplot(3,3,7)
plt.plot(newdf["Month"], newdf["p_age"])
plt.title("accumulated age of president")

plt.subplot(3,3,8)
plt.plot(newdf["Month"], newdf["Temperature"])
plt.title("Temperature")

plt.show()
```

```python
import statsmodels.tsa.stattools as stattools

partial1 = stattools.pacf(newdf["log_S&P"], nlags = 32)
partial2 = stattools.pacf(newdf["yield"], nlags = 32)
partial3 = stattools.pacf(newdf["log_earnings"], nlags = 32)
partial4 = stattools.pacf(newdf["log_CPI"], nlags = 32)
partial5 = stattools.pacf(newdf["10yr_rate"], nlags = 32)
partial6 = stattools.pacf(newdf["log_gold"], nlags = 32)
partial7 = stattools.pacf(newdf["p_age"], nlags = 32)
partial8 = stattools.pacf(newdf["Temperature"], nlags = 32)

plt.figure(figsize=(18,15))
plt.subplot(3,3,1)
plt.plot(partial1, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, $\log(S&T)$")

plt.subplot(3,3,2)
plt.plot(partial2, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, Yield")

plt.subplot(3,3,3)
plt.plot(partial3, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, log(earnings)")

plt.subplot(3,3,4)
plt.plot(partial4, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, log(CPI)")

plt.subplot(3,3,5)
plt.plot(partial5, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, 10 Year Rate")

plt.subplot(3,3,6)
plt.plot(partial6, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, log(gold)")

plt.subplot(3,3,7)
plt.plot(partial7, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, accumulated age of president")

plt.subplot(3,3,8)
plt.plot(partial8, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, Temperature")

plt.show()
```

It looks like all variables have very high correlation with the first lagged term $u_{t-1}.$

All the trend variables (S&T, earning, CPI, gold price) have little correlation with higher order lagged terms.

The rate variables (yield, 10 yr rate) and temperature have also correlation with higher order lagged terms.

# Question 4, Regress the month t S&P500 price on all eight variables from month t-1. Comment briefly on the results. Look at the t-statistics of the coefficients, including the constant term, and also consider the economic significance of the size of the coefficients. What do standard regression diagnostics say about the quality of the fit?

```python
df_for4 = newdf[["yield","log_earnings", "log_CPI", "10yr_rate","log_gold","p_age","Temperature"]][:-1].copy()
df_for4["log_S&P"] = newdf["log_S&P"][1:].values
df_for4.tail()
```

```python
df_for4 = sm.add_constant(df_for4)
xvars4 = ["yield","log_earnings", "log_CPI", "10yr_rate","log_gold","p_age","Temperature","const"]
reg4 = sm.OLS(endog=df_for4["log_S&P"], exog=df_for4[xvars4]).fit()
reg4.summary()
```

In general a very good fit, with $R^2 = 99\%.$ Yield is the most significant factor, with a coeff of -229. Earnings, CPI and 10yr_rate are also crutial.

    Low yield          ---> high stock price
    High earnings      ---> high stock price
    High CPI           ---> high stock price
    Low interest rate  ---> high stock price

```python
predicted4 = reg4.predict(df_for4[xvars4])
plt.plot(np.linspace(1880, 2018, len(df_for4)), predicted4, label="Fitted values")
plt.plot(np.linspace(1880, 2018, len(df_for4)), df_for4["log_S&P"], label="Actual values")
plt.title("Actual v. fitted log S&P")
plt.legend()
plt.show()
```

```python
#t1 = np.insert(newdf["log_S&P"].values, 0, np.nan)[:-1]
newdf["t1"] = np.insert(newdf["log_S&P"].values, 0, np.nan)[:-1]
newdf["t2"] = np.insert(newdf["log_S&P"].values, 0, [np.nan, np.nan])[:-2]
newdf["t3"] = np.insert(newdf["log_S&P"].values, 0, [np.nan, np.nan, np.nan])[:-3]
newdf["t4"] = np.insert(newdf["log_S&P"].values, 0, [np.nan, np.nan, np.nan, np.nan])[:-4]
newdf["t5"] = np.insert(newdf["log_S&P"].values, 0, [np.nan, np.nan, np.nan, np.nan, np.nan])[:-5]
newdf["t6"] = np.insert(newdf["log_S&P"].values, 0, [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])[:-6]
```

```python
df_for8 = newdf.loc[6:]
df_for8 = df_for8.reset_index(drop = True)
xvars8 = ["t1", "t2", "t3", "t4", "t5", "t6"]
reg8 = sm.OLS(endog=df_for8["log_S&P"], exog=df_for8[xvars8]).fit()
reg8.summary()
```

```python
predicted8 = reg8.predict(df_for8[xvars8])
plt.plot(np.linspace(1880, 2018, len(df_for8)), predicted8, label = "Fitted value")
plt.plot(np.linspace(1880, 2018, len(df_for8)), df_for8["log_S&P"],"--", label = "Actual value")
plt.legend()
plt.title("Log S&P, regress on lagged values")
plt.show()
```

```python
residual_series = df_for8["log_S&P"].values - predicted8

plt.figure(figsize=(10,6))
plt.subplot(1,2,1)

plt.plot(np.linspace(1880, 2018, len(df_for8)), residual_series)
plt.title("Residuals")

plt.subplot(1,2,2)
partial_res = stattools.pacf(residual_series, nlags = 32)
plt.plot(partial_res, "-o", markersize = 3)
plt.title("Partial autoregression")
```
