```python
import math
import numpy as np
import numpy.linalg as la
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
raw_data = pd.read_csv("data_file2.csv")
```

```python
raw_data["Month"] = pd.to_datetime(raw_data["Month"])
raw_data
```

# Question 1, transform the data

### 1, Take the natural logarithm of the S&P500 price

```python
newdf = pd.DataFrame(raw_data["Month"])
newdf["log_S&P"]=np.log(raw_data["S&P500"])
```

### 2, Compute the dividend yield as dividend divided by price

```python
newdf["yield"] = raw_data["Dividends"] / raw_data["S&P500"]
```

### 3, Take the natural logarithm of earnings

```python
newdf["log_earnings"] = np.log(raw_data["Earnings"])
```

### 4, Take the natural logarithm of CPI

```python
newdf["log_CPI"] = np.log(raw_data["CPI"])
```

### 5, Leave the 10-year rate untransformed

```python
newdf["10yr_rate"] = raw_data["10-year rate"]
```

### 6, Take the natural logarithm of the price of gold

```python
newdf["log_gold"] = np.log(raw_data["Gold"])
```

### 7, Accumulate the President's age, that is each month take the previous month's value (start at zero) and add the current month's value

```python
accumulate_age = np.array(raw_data["President age"])[1:] + np.array(raw_data["President age"])[:-1]
newdf["p_age"] = np.insert(accumulate_age, 0, raw_data["President age"][0])
```

### 8, Leave the temperature untransformed

```python
newdf["Temperature"] = raw_data["Temperature"]
```

```python
import statsmodels.tsa.stattools as stattools

partial1 = stattools.pacf(newdf["log_S&P"], nlags = 32)
partial2 = stattools.pacf(newdf["yield"], nlags = 32)
partial3 = stattools.pacf(newdf["log_earnings"], nlags = 32)
partial4 = stattools.pacf(newdf["log_CPI"], nlags = 32)
partial5 = stattools.pacf(newdf["10yr_rate"], nlags = 32)
partial6 = stattools.pacf(newdf["log_gold"], nlags = 32)
partial7 = stattools.pacf(newdf["p_age"], nlags = 32)
partial8 = stattools.pacf(newdf["Temperature"], nlags = 32)

plt.figure(figsize=(18,15))
plt.subplot(3,3,1)
plt.plot(partial1, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, $\log(S&T)$")

plt.subplot(3,3,2)
plt.plot(partial2, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, Yield")

plt.subplot(3,3,3)
plt.plot(partial3, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, log(earnings)")

plt.subplot(3,3,4)
plt.plot(partial4, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, log(CPI)")

plt.subplot(3,3,5)
plt.plot(partial5, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, 10 Year Rate")

plt.subplot(3,3,6)
plt.plot(partial6, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, log(gold)")

plt.subplot(3,3,7)
plt.plot(partial7, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, accumulated age of president")

plt.subplot(3,3,8)
plt.plot(partial8, "-o", markersize = 3.5)
plt.title("Partial autocorrelation, Temperature")

plt.show()
```

# Question 4, Regress the month t S&P500 price on all eight variables from month t-1. Comment briefly on the results. Look at the t-statistics of the coefficients, including the constant term, and also consider the economic significance of the size of the coefficients. What do standard regression diagnostics say about the quality of the fit?

```python
df_for4 = newdf[["yield","log_earnings", "log_CPI", "10yr_rate","log_gold","p_age","Temperature"]][:-1].copy()
df_for4["log_S&P"] = newdf["log_S&P"][1:].values
df_for4.tail()
```

```python
df_for4 = sm.add_constant(df_for4)
xvars4 = ["yield","log_earnings", "log_CPI", "10yr_rate","log_gold","p_age","Temperature","const"]
reg4 = sm.OLS(endog=df_for4["log_S&P"], exog=df_for4[xvars4]).fit()
reg4.summary()
```

# 5, Repeat 4 for the price of gold in month t as the dependent variable.

```python
df_for5 = newdf[["yield","log_earnings", "log_CPI", "10yr_rate","log_S&P","p_age","Temperature"]][:-1].copy()
df_for5["log_gold"] = newdf["log_gold"][1:].values

df_for5 = sm.add_constant(df_for5)
xvars5 = ["yield","log_earnings", "log_CPI", "10yr_rate","log_S&P","p_age","Temperature","const"]
reg5 = sm.OLS(endog=df_for5["log_gold"], exog=df_for5[xvars5]).fit()
reg5.summary()
```

# 6, Repeat 5 for the dividend yield as the dependent variable.

```python
df_for6 = newdf[["log_gold","log_earnings", "log_CPI", "10yr_rate","log_S&P","p_age","Temperature"]][:-1].copy()
df_for6["yield"] = newdf["yield"][1:].values

df_for6 = sm.add_constant(df_for6)
xvars6 = ["log_gold","log_earnings", "log_CPI", "10yr_rate","log_S&P","p_age","Temperature","const"]
reg6 = sm.OLS(endog=df_for6["yield"], exog=df_for6[xvars6]).fit()
reg6.summary()
```

# 7, Use your regression to predict in-sample each month's S&P500 value, dividend yield and gold price. Compute the return on a portfolio that invests each month in whichever is higher, the predicted return of the S&P500 (capital appreciation plus dividend yield) or the predicted return of gold (price appreciation only). Compare that to investing in the S&P500 every month, and investing in gold every month.

```python
# predict in-sample S&P 500, using reg4
plt.figure(figsize=(8,5))
predicted_SP = reg4.predict(df_for4[xvars4])
xaxis = np.linspace(1880, 2018, len(df_for4))
plt.plot(xaxis, predicted_SP, label = "in-sample prediction")
plt.plot(xaxis, df_for4["log_S&P"], label = "actual value")
plt.legend()
plt.title("$\log$(S&P), actual v. predicted")
plt.show()
```

```python
# predict in-sample yield, using reg6
plt.figure(figsize=(8,5))
predicted_yield = reg6.predict(df_for6[xvars6])
xaxis = np.linspace(1880, 2018, len(df_for6))
plt.plot(xaxis, predicted_yield, label = "in-sample prediction")
plt.plot(xaxis, df_for6["yield"], label = "actual value")
plt.legend()
plt.title("yield, actual v. predicted")
plt.show()
```

```python
# predict in-sample gold price, using reg5
plt.figure(figsize=(8,5))
predicted_gold = reg5.predict(df_for5[xvars5])
xaxis = np.linspace(1880, 2018, len(df_for5))
plt.plot(xaxis, predicted_gold, label = "in-sample prediction")
plt.plot(xaxis, df_for5["log_gold"], label = "actual value")
plt.legend()
plt.title("$\log$(gold), actual v. predicted")
plt.show()
```
